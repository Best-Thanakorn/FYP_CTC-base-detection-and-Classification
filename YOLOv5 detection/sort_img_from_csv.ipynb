### Copy wanted images based on .csv

```python
import os
import csv
import shutil

# Path to the CSV file containing the file names
csv_file_path = r"D:\FYP\CTC_CCA_dataset\CTC_CCA_dataset\CTC_cell_dectection - Copy.csv"

# Source directory containing the files to be copied
source_dir = r"D:\FYP\CTC_CCA_dataset\CTC_CCA_dataset\raw_images_for_model\brightfield"

# Destination directory where files will be copied
destination_dir = r"D:\FYP\CTC_CCA_dataset\CTC_CCA_dataset\brightfield_for_train"

# Create the destination directory if it doesn't exist
if not os.path.exists(destination_dir):
    os.makedirs(destination_dir)

# Read the CSV file and copy files based on the file names
with open(csv_file_path, 'r') as file:
    csv_reader = csv.reader(file)
    for row in csv_reader:
        file_name = row[0]  # Assuming file names are in the first column
        source_file_path = os.path.join(source_dir, file_name)
        if os.path.exists(source_file_path):
            destination_file_path = os.path.join(destination_dir, file_name)
            shutil.copy2(source_file_path, destination_file_path)
            print(f"File '{file_name}' copied successfully.")

print("All files copied successfully.")
```
