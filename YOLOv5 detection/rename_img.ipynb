```python
import os

def rename_files_in_folder(folder_path):
    # Get all files in the specified folder
    files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
    
    # Sort files to maintain order
    files.sort()
    
    for index, filename in enumerate(files):
        # Split the filename into name and extension
        name, ext = os.path.splitext(filename)
        
        # Create new filename with the specified format
        new_filename = f"{index}_{name}{ext}"
        
        # Construct full file paths
        original_file = os.path.join(folder_path, filename)
        new_file = os.path.join(folder_path, new_filename)
        
        # Rename the file
        os.rename(original_file, new_file)
        print(f'Renamed: {original_file} to {new_file}')

# Example usage
folder_path = r"D:\FYP\CTC_CCA_dataset\CTC_CCA_dataset\raw_images_for_model\fluorescence_png - Copy - Copy"  # Replace with your folder path
rename_files_in_folder(folder_path)
```

```python
import os

def compare_folders(folder1, folder2):
    # Get all files in both folders
    files1 = set(os.listdir(folder1))
    files2 = set(os.listdir(folder2))
    
    # Find files that are in folder1 but not in folder2
    only_in_folder1 = files1 - files2
    
    # Find files that are in folder2 but not in folder1
    only_in_folder2 = files2 - files1
    
    # Print results
    if only_in_folder1:
        print(f"Files only in {folder1}:")
        for file in only_in_folder1:
            print(file)
    else:
        print(f"No unique files in {folder1}.")

    if only_in_folder2:
        print(f"\nFiles only in {folder2}:")
        for file in only_in_folder2:
            print(file)
    else:
        print(f"No unique files in {folder2}.")

# Example usage
folder1_path = r"D:\FYP\CTC_CCA_dataset\CTC_CCA_dataset\Annotations\mix"  # Replace with your first folder path
folder2_path = r"D:\FYP\CTC_CCA_dataset\CTC_CCA_dataset\raw_images_for_model\brightfield_png"  # Replace with your second folder path
compare_folders(folder1_path, folder2_path)
```

```python
import os

def find_unmatched_files(folder1, folder2):
    # Get all files in both folders
    files1 = set(os.listdir(folder1))
    files2 = set(os.listdir(folder2))

    # Create sets for just the base filenames (without extensions)
    base_filenames1 = {os.path.splitext(f)[0] for f in files1}
    base_filenames2 = {os.path.splitext(f)[0] for f in files2}

    # Find unmatched files
    unmatched_in_folder1 = base_filenames1 - base_filenames2
    unmatched_in_folder2 = base_filenames2 - base_filenames1

    # Print results
    if unmatched_in_folder1:
        print(f"Files only in {folder1}:")
        for filename in unmatched_in_folder1:
            print(f"{filename}.png")  # Assuming the missing files are .png
    else:
        print(f"No unique files in {folder1}.")

    if unmatched_in_folder2:
        print(f"\nFiles only in {folder2}:")
        for filename in unmatched_in_folder2:
            print(f"{filename}.xml")  # Assuming the missing files are .xml
    else:
        print(f"No unique files in {folder2}.")

# Example usage
folder1_path = r"D:\FYP\CTC_CCA_dataset\CTC_CCA_dataset\Annotations\mix"  # Replace with your first folder path
folder2_path = r"D:\FYP\CTC_CCA_dataset\CTC_CCA_dataset\raw_images_for_model\brightfield_png"  # Replace with your second folder path
find_unmatched_files(folder1_path, folder2_path)
```

```python
import os

def find_missing_files(folder1, folder2):
    # Get all files in both folders
    files1 = set(os.listdir(folder1))
    files2 = set(os.listdir(folder2))

    # Extract the base filenames (without extensions)
    base_filenames1 = {os.path.splitext(f)[0] for f in files1}
    base_filenames2 = {os.path.splitext(f)[0] for f in files2}

    # Find the missing file
    missing_file = base_filenames2 - base_filenames1

    if missing_file:
        print(f"The missing file is: {missing_file.pop()}.xml")
    else:
        print("No missing files found.")

# Example usage
folder1_path = r"D:\FYP\CTC_CCA_dataset\CTC_CCA_dataset\Annotations\mix"  # Replace with your first folder path
folder2_path = r"D:\FYP\CTC_CCA_dataset\CTC_CCA_dataset\raw_images_for_model\brightfield_png"  # Replace with your second folder path
find_missing_files(folder1_path, folder2_path)
```

```python
import os

def rename_files(folder_path):
    for filename in os.listdir(folder_path):
        if filename.endswith('.png'):
            # Split the filename by the underscore
            parts = filename.split('_')
            
            # Construct the new filename without the indexing number
            new_filename = '_'.join(parts[1:])
            
            # Construct the full paths for the old and new filenames
            old_filepath = os.path.join(folder_path, filename)
            new_filepath = os.path.join(folder_path, new_filename)
            
            # Rename the file
            os.rename(old_filepath, new_filepath)
            print(f'Renamed: {old_filepath} to {new_filepath}')

# Example usage
folder_path = r'D:\FYP\CTC_CCA_dataset\CTC_CCA_dataset\raw_images_for_model\brightfield_png'  # Replace with the path to your folder
rename_files(folder_path)
```
